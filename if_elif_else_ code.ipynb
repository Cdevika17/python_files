```python
#Triangle checker

x=int(input("side1:"))
y=int(input("side2:"))
z=int(input("side3:"))

if x+y>z and y+z>x and z+x>y:
    print("Triangle is formed")
    if x==y==z:
      print("It is a Equaliteral Triangle")
    elif x==y or y==z or z==x :
       print("It is a Isoslate triangle")
    elif x!=y!=z:
      print("It is a Scalen Triangle")
else:
  print("Triangle is not formed")
```

```python
# Employee Bonus System

emp_salary=int(input("Enter your salary:"))
years_of_service=int(input("Enter your year of service:"))


if years_of_service >=10:
    Bonus=(emp_salary*20)/100

elif years_of_service >=5:
    Bonus=(emp_salary*10)/100

else:
    Bonus=(emp_salary*5)/100

print("Total salary is:{}".format(emp_salary+Bonus))

if emp_salary > 50000:
   Bonus=Bonus+(emp_salary*5)/100
   print("extra bonus with salary is:{}".format(emp_salary+Bonus))
```
